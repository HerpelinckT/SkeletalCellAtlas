```r
library(Seurat)
library(ggplot2)
library(patchwork)
```

```r
gp <- readRDS(file = "../data/growth_plate_visualization.Rds")
gp
```

```r
colnames(gp[[]])
```

```r
# gene or list of genes of interest
gois <- c("Sox9", "Runx2", "Col2a1", "Col10a1", "Ihh")
```

```r
ggobjs = RidgePlot(gp, features = gois, ncol = 1, group.by='CellType')
```

```r
options(repr.plot.width=5, repr.plot.height=15, repr.plot.res = 400)
ggobjs
```
